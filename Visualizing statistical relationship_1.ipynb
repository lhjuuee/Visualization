## Scatter plots

```python
sns.relplot(x='play_time', y='npc_hongmun', data = train_play_time)
```

#### add 'hue' parameter

```python
sns.relplot(x='play_time', y='npc_hongmun', hue='label', data = train_play_time)
```

#### add 'style' parameter

```python
sns.relplot(x='play_time', y='npc_hongmun', hue ='label', style='label', data = train_play_time)
```

#### add 'palette' parameter

```python
sns.relplot(x='play_time', y='npc_hongmun', hue='label', palette="ch:r=-.5,l=.75", data=train_play_time)
```

#### add 'size' parameter

```python
sns.relplot(x='play_time', y='npc_hongmun', size='label', data=train_play_time)
```

#### Range can be customized by using 'sizes'

```python
sns.relplot(x='play_time', y='npc_hongmun', size='label', sizes=(10,100), data=train_play_time)
```
