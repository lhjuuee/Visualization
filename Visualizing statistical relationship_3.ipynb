```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
from scipy import stats
```

# Plotting subsets of data with semantic mappings

```python
day2 = np.random.randint(1,20, size=100)
day2.sort()
value2 = np.random.randn(100).cumsum()
df2 = pd.DataFrame({'day': day2, 'value':value2})
x=np.random.randint(0,3, size=100)
df2['label'] = x
```

```python
sns.relplot(x = 'day', y='value', hue='label', kind='line',data=df2, estimator=None)
```

```python
mapping = {0:'Korea', 1:'America', 2:'Germany'}
df2['Nation'] = df2['label'].map(mapping)
x=np.random.randint(0,3, size=100)
df2['label'] = x
```

```python
sns.relplot(x='day', y='value', hue='label', style='Nation', kind='line',
           data=df2)
```

#### Dashes, Markers (style)

```python
sns.relplot(x='day', y='value', hue='label', style='Nation', kind='line',
           dashes=False, markers=True, data=df2)
```

#### Color value. palette

```python
palette = sns.cubehelix_palette(light=.9, n_colors=3)
sns.relplot(x='day', y='value', hue='label', style='Nation', kind='line',
           palette=palette, data=df2)
```

#### add 'size parameter

```python
sns.relplot(x='day', y='value', hue = 'label', size='Nation',
           kind='line', estimator=None, data = df2)
```

#### Or, You can see those plot as lattice plot or small-multiples

```python
sns.relplot(x='day', y='value', hue = 'label',
            col='Nation', col_wrap=2,
            height=10, aspect=.9, linewidth=2.5,
           kind='line', data = df2)
```
